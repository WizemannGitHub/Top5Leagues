```python
import matplotlib.pyplot as plt
import pandas as pd
from ipywidgets import interact, widgets
```

```python
%matplotlib inline
```

```python
data = pd.read_csv('footballdata.xls')
```

```python
goalsPerPlayer = data[['Goals','Player']]
```

```python
goalsPerPlayer
```

```python
#Top10 Goalscorers in Top5 Leagues

goalsPerPlayerSorted = goalsPerPlayer.sort_values(by='Goals', ascending=False).head(10)
plt.figure(figsize=(20, 6))
plt.bar(goalsPerPlayerSorted['Player'], goalsPerPlayerSorted['Goals'])
plt.xlabel('Spieler')
plt.ylabel('Tore')
plt.title('Tore pro Spieler')
```

```python
def update_plot(selected_player):
    plt.figure(figsize=(20, 6))
    plt.bar(goalsPerPlayerSorted['Player'], goalsPerPlayerSorted['Goals'], color='lightblue')
    plt.bar(goalsPerPlayerSorted[goalsPerPlayerSorted['Player'] == selected_player]['Player'],
            goalsPerPlayerSorted[goalsPerPlayerSorted['Player'] == selected_player]['Goals'], color='orange')
    plt.xlabel('Player')
    plt.ylabel('Goals')
    plt.title(f'{selected_player} - Goals')
    plt.show()

# Create a dropdown menu with player names
player_dropdown = widgets.Dropdown(options=goalsPerPlayerSorted['Player'].tolist(), description='Player:')

# Set up the interaction between the dropdown and the update function
interact(update_plot, selected_player=player_dropdown)
```
