```python
import pandas as pd
from pathlib import Path
```

```python
season = '2022-2023'

url_standard = 'https://fbref.com/en/comps/Big5/' + season + '/stats/players/' + season + '-Big-5-European-Leagues-Stats'
```

```python
df = pd.read_html(url_standard)[0]
```

```python
df
```

```python
df_copy = df.copy()
```

```python
df_copy.columns = df_copy.columns.droplevel()
```

```python
df_copy = df_copy.rename(columns={'Rk' : 'ID'})
```

```python
df_copy.rename(columns={'Pos':'Position'}, inplace = True)
```

```python
df_copy['Position'] = df_copy['Position'].str[:2]
df_copy
```

```python
df = df_copy
```

```python
df_copy = df
```

```python
df_copy['Nation'] = df_copy['Nation'].str.split(n=1).str[1]
```

```python
df_copy.rename(columns={'Comp': 'League'}, inplace=True)
```

```python
df_copy['League'] = df_copy['League'].str.split(n=1).str[1]
```

```python
df_copy.columns
```

```python
df_copy.drop('90s', axis=1, inplace=True)
```

```python
columns_to_drop = ['PK','PKatt', 'xG', 'npxG', 'xAG', 'npxG+xAG', 'PrgC', 'PrgP', 'PrgR', 'G+A-PK', 'xG', 'xAG', 'xG+xAG', 'npxG', 'npxG+xAG']
df_copy.drop(columns=columns_to_drop, axis=1, inplace=True)
```

```python
df_copy
```

```python
df.columns.values[18] = 'G+A per Game'
```

```python
df_copy.drop('Matches', axis=1, inplace=True)
```

```python
df_copy.drop(df_copy[df_copy['ID'] == 'Rk'].index, inplace=True)
```
